```python

%matplotlib inline 
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.patches as patches

matplotlib.rcParams['savefig.dpi'] = 2 * matplotlib.rcParams['savefig.dpi']
```

```python
counts = [i/10 for i in list(range(1,50)) if i/10 >= 1]


vals = [1/i for i in counts]

with plt.xkcd():
    
    fig = plt.figure(figsize=(8,5))
    
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    plt.plot(counts,vals)

    plt.ylabel('Amount of feature engineering')
    plt.xlabel('Amount of training data')
    
    plt.tick_params(
    axis='both',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    left='off',# ticks along the top edge are off
    labelbottom='off',
    labelleft='off') # labels along the bottom edge are off
    
    plt.title('BUILDING A ROBUST MODEL',y=1.08)
    
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    
    
    plt.annotate(
    'PURE SOFTWARE ENGINEERING',
    xy=(1, 1), arrowprops=dict(facecolor='black', shrink=0.09), xytext=(1.5, 0.8))
    
    plt.annotate(
    'UN-STRUCTURED DATA IN \n DEEP NEURAL NETWORK',
    xy=(5, 0.2), arrowprops=dict(facecolor='black', shrink=0.1), xytext=(2.7, 0.45))
    
    plt.xlim(1,5)
    

    
    plt.show()
```

```python
import random
```

```python
x = list(range(1,7))

random.seed(46)
y = [(i + random.random()*2)*10 for i in x]
with plt.xkcd():
    fig = plt.figure(figsize=(8,4))
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    plt.scatter(x,y, s=280)
    
    plt.ylim(0,80)
    plt.xlim(0,8)
    
    plt.xlabel('HEIGHT')
    plt.ylabel('WEIGHT')
    
    plt.title('HEIGHT TO WEIGHT DATA',y=1.08)
    
    plt.show()
```

```python
with plt.xkcd():
    fig = plt.figure(figsize=(8,4))
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    plt.plot([0.5,6.6],[18,73], ls="dashed",lw=3, c='r', zorder=-1)
    plt.scatter(x,y, s=280, alpha=0.1)
    
    plt.ylim(0,80)
    plt.xlim(0,8)
    
    plt.xlabel('HEIGHT')
    plt.ylabel('WEIGHT')
    
    ax.annotate('WEIGHT = 9 x HEIGHT + 15', xy=(1, 1), xytext=(1.7, 64),rotation=21
#             arrowprops=dict(facecolor='black', shrink=0.05),
            )
    plt.title('MODEL BASED LEARNING',y=1.08)
    
    plt.scatter([4.5],[54], marker='.', s=1550, c='orange')
    
    plt.annotate(
    'Weight prediction for the height of 4.5 = 55.5\n (9 x 4.5 + 15 = 55.5)',
    xy=(4.55, 50), xytext=(2.5, 10),
    arrowprops=dict(facecolor='black', shrink=0.09),)

    plt.show()
```
